###                        ***KÜTÜPHANE KATALOĞU OTOMASYON UYGULAMASI***

```python
dosya = "/content/drive/MyDrive/Colab Notebooks/HU -BBY162 -2021/final.txt"

print("Kütüphane Kataloğu Otomasyonuna Hoşgeldiniz...")

def eserListele():
  f = open(dosya, 'r')
  for line in f.readlines():
    print(line)
  f.close()

def yeniEserEkle():
    print("Yeni Eser Ekle")
    eserAdi = input("Eser Adı Giriniz:")
    yazarAdi = input("Yazar Adı Giriniz:")
    yayinEvi = input("Yayınevini Giriniz:")
    basimTarihi = input("Basım Tarihini Giriniz:")
    ISBN = input("ISBN Numarasını Giriniz:")
    f = open(dosya, "a")
    f.write(eserAdi + " , " + yazarAdi + " , " + yayinEvi + " | " + basimTarihi + " | " + ISBN + " " + "\n")
    print("Kayıt Başarılı Bir Şekilde Eklendi")
    f.close()


    
def menu():

    dosyaListesi=[yeniEserEkle, eserListele]
    while True:
        print("1- Yeni Eser Ekle")
        print("2- Kayıt Listele")
        print("0-Programdan Çık")
        seciminiz=int(input("Lütfen Yapmak İstediğiniz İşlem Seçiniz (0-2):"))
        print("\n")
        if seciminiz<=2 and seciminiz>=1:
            dosyaListesi[seciminiz-1]()

            print("\n" * 2)
        elif seciminiz==0:
            print("Çıkış Yapılıyor")
            break
        else:
            print("Lütfen Seçim İçin 0 ile 2 arası bir rakam giriniz!")



menu()
```
